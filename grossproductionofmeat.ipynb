```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import math
from statistics import mean,variance, stdev, mode, median
import pandas
```

```python
data = pandas.read_csv("gross.csv")
data.set_index('country', inplace=True)
data= data.sort_values('2010')


data
```

```python
for year in range(2010,2019):
    data.plot.bar(y=str(year), figsize=(12,12), title=str(year))
    plt.figure(figsize=(50,50))
```

```python
vertical=['2010','2011','2012','2013','2014','2015','2016','2017','2018','2019']
data[vertical].mean()
```

```python
data[vertical].std()
```

```python
data[vertical].var()
```

```python
data[vertical].mode
```

```python
data[vertical].median
```

```python
data.groupby('country').median().plot.bar(stacked=True)
plt.title('median')
plt.show()
```

```python
for year in range(2010,2019):
    data.plot.bar(y=str(year),figsize=(11,11),title=str(year))
    plt.figure(figsize=(42,42))
```

```python
for year in range(2010,2019):
    data.plot.hist(y=str(year),figsize=(11,11),title=str(year))
    plt.figure(figsize=(42,42))
```

```python
for year in range(2010,2019):
    data.plot.line(y=str(year),figsize=(11,11),title=str(year))
    plt.figure(figsize=(42,42))
```
